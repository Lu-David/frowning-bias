```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import json
```

```python
# load ../results/baseline/baseline_train_log.json
with open('../baseline_train_log.json') as f:
    baseline_train_log = json.load(f)
```

```python
baseline_train_log
```

```python
# plot training and validation loss
plt.plot(baseline_train_log['loss']['train'], label='train')
plt.plot(baseline_train_log['loss']['val'], label='validation')
plt.legend()
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Baseline Training and Validation Loss')
```

```python
# plot training and validation accuracy
plt.plot(baseline_train_log['acc']['train'], label='train')
plt.plot(baseline_train_log['acc']['val'], label='validation')
plt.legend()
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Baseline Training and Validation Accuracy')
```

```python
# plot training and validation auc
plt.plot(baseline_train_log['auc']['train'], label='train')
plt.plot(baseline_train_log['auc']['val'], label='validation')
plt.legend()
plt.xlabel('Epoch')
plt.ylabel('AUC')
plt.title('Baseline Training and Validation AUC')
```
